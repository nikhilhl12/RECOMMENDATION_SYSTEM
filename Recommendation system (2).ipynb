```python
user_data = data1.pivot_table(index='UserID',columns='BookTitle',values='BookRating')
```

```python
user_data.values
```

```python
user_data.fillna(0,inplace=True)
```

```python
user_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_data.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_data = pd.DataFrame(user_sim)
```

```python
user_sim_data.shape
```

```python
user_sim_data.index = user_data.index
user_sim_data.columns = user_data.index
```

```python
np.fill_diagonal(user_sim,0)
user_sim_data.iloc[0:5,0:5]
```

```python
def get_reco(UserID):
    tem = list(user_sim_data.sort_values([UserID],ascending=False).head(100).index)
    book_list = list(data1[data1["UserID"].isin(tem)]["BookTitle"])
    #print(len(book_list)
    return set(book_list)-set(data1[data1["UserID"]==UserID]["BookTitle"])
    
```
